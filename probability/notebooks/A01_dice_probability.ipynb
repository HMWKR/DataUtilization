## 기본 용어 정리

| 용어 | 영문 | 정의 |
|------|------|------|
| **시행(실험)** | Experiment | 결과가 불확실한 관찰이나 행위 |
| **표본공간** | Sample Space ($S$ 또는 $\Omega$) | 시행에서 가능한 모든 결과의 집합 |
| **근원사건** | Elementary Event | 표본공간의 각 원소. 더 이상 분해할 수 없는 기본 결과 |
| **사건** | Event | 표본공간의 부분집합. 관심 있는 결과의 모임 |
| **여사건** | Complement ($A^c$) | 사건 $A$가 일어나지 않는 사건. $P(A^c) = 1 - P(A)$ |
| **공사건** | Empty Event ($\emptyset$) | 절대 일어나지 않는 사건 |
| **전사건** | Certain Event ($S$) | 반드시 일어나는 사건 = 표본공간 자체 |

### 고전적 확률 정의

모든 근원사건이 **동일한 확률(equally likely)**을 가질 때:

$$P(A) = \frac{\text{사건 } A \text{에 포함된 근원사건의 수}}{\text{표본공간의 전체 근원사건의 수}} = \frac{|A|}{|S|}$$

### 확률의 공리 (Kolmogorov Axioms)

1. **비음수성**: $P(A) \geq 0$ (어떤 사건의 확률도 0 이상)
2. **정규성**: $P(S) = 1$ (표본공간 전체의 확률은 1)
3. **가산 가법성**: 서로 배반인 사건 $A_1, A_2, \ldots$에 대해 $P(\bigcup A_i) = \sum P(A_i)$

## 문제 1: 1개 주사위를 던져서 5가 나올 확률

### 문제 정의

공정한(fair) 주사위 1개를 던질 때, 눈금 5가 나올 확률을 구하라.

### 수학적 풀이

**표본공간 정의:**

$$S = \{1, 2, 3, 4, 5, 6\}, \quad |S| = 6$$

**사건 정의:**

$$A = \{5\}, \quad |A| = 1$$

**확률 계산:**

$$P(X = 5) = \frac{|A|}{|S|} = \frac{1}{6} \approx 0.1667$$

공정한 주사위에서 각 면이 나올 확률은 모두 동일하게 $\frac{1}{6}$이다.  
이는 **고전적 확률 정의**를 직접 적용한 결과이다.

```python
# 문제 1: P(X=5)
sample_space = {1, 2, 3, 4, 5, 6}
event_A = {5}

P_A = len(event_A) / len(sample_space)

print("=" * 50)
print("문제 1: 주사위를 던져서 5가 나올 확률")
print("=" * 50)
print(f"표본공간 S = {sample_space}")
print(f"|S| = {len(sample_space)}")
print(f"사건 A = {event_A}")
print(f"|A| = {len(event_A)}")
print(f"\nP(X=5) = {len(event_A)}/{len(sample_space)} = {P_A:.4f}")
print(f"분수 표현: 1/6 ≈ {1/6:.6f}")
```

```python
# 문제 1 시각화: 각 눈금의 확률
faces = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
colors = ['#66c2a5' if f != 5 else '#fc8d62' for f in faces]

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(faces, probabilities, color=colors, edgecolor='white', linewidth=2, width=0.6)

# 확률값 표시
for bar, p in zip(bars, probabilities):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
            f'{p:.4f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

ax.set_xlabel('주사위 눈금', fontsize=14)
ax.set_ylabel('확률 P(X=k)', fontsize=14)
ax.set_title('공정한 주사위: 각 눈금의 확률 (이산 균등분포)', fontsize=16, fontweight='bold')
ax.set_xticks(faces)
ax.set_ylim(0, 0.25)
ax.axhline(y=1/6, color='red', linestyle='--', alpha=0.5, label=f'P = 1/6 ≈ {1/6:.4f}')
ax.legend(fontsize=12)

# 5번 눈금 강조 텍스트
ax.annotate('P(X=5) = 1/6', xy=(5, 1/6), xytext=(5.5, 0.22),
            fontsize=13, fontweight='bold', color='#fc8d62',
            arrowprops=dict(arrowstyle='->', color='#fc8d62', lw=2))

plt.tight_layout()
plt.show()
```

### 해석

- 공정한 주사위는 **이산 균등분포(Discrete Uniform Distribution)**를 따른다.
- 6개의 근원사건 각각이 동일한 확률 $\frac{1}{6}$을 가진다.
- 특정 눈금(여기서는 5)이 나올 확률은 $\frac{1}{6} \approx 16.67\%$이다.
- 이것은 **선험적(a priori) 확률**: 실험 전에 논리적 추론으로 결정할 수 있다.

## 문제 2: 1개 주사위를 던져서 5 이하가 될 확률

### 문제 정의

공정한 주사위 1개를 던질 때, 눈금이 5 이하가 될 확률을 구하라.

### 수학적 풀이 — 방법 1: 직접 계산

**사건 정의:**

$$B = \{X \leq 5\} = \{1, 2, 3, 4, 5\}, \quad |B| = 5$$

**확률 계산:**

$$P(X \leq 5) = \frac{|B|}{|S|} = \frac{5}{6} \approx 0.8333$$

### 수학적 풀이 — 방법 2: 여사건(Complement) 활용

"5 이하"의 여사건은 "6"이다:

$$B^c = \{X > 5\} = \{6\}, \quad |B^c| = 1$$

$$P(B^c) = \frac{1}{6}$$

여사건의 성질을 이용하면:

$$P(B) = 1 - P(B^c) = 1 - \frac{1}{6} = \frac{5}{6} \approx 0.8333$$

> **여사건 활용의 장점**: 사건 자체의 경우의 수가 많을 때, 여사건(경우의 수가 적음)을 이용하면 계산이 간편해진다.

```python
# 문제 2: P(X <= 5)
event_B = {1, 2, 3, 4, 5}
event_B_complement = {6}

# 방법 1: 직접 계산
P_B_direct = len(event_B) / len(sample_space)

# 방법 2: 여사건 활용
P_B_complement = len(event_B_complement) / len(sample_space)
P_B_via_complement = 1 - P_B_complement

print("=" * 50)
print("문제 2: 주사위를 던져서 5 이하가 나올 확률")
print("=" * 50)

print("\n[ 방법 1: 직접 계산 ]")
print(f"사건 B = {event_B}")
print(f"|B| = {len(event_B)}")
print(f"P(X≤5) = {len(event_B)}/{len(sample_space)} = {P_B_direct:.4f}")

print("\n[ 방법 2: 여사건 활용 ]")
print(f"여사건 B^c = {event_B_complement}")
print(f"P(B^c) = P(X>5) = {len(event_B_complement)}/{len(sample_space)} = {P_B_complement:.4f}")
print(f"P(B) = 1 - P(B^c) = 1 - {P_B_complement:.4f} = {P_B_via_complement:.4f}")

print(f"\n두 방법의 결과 일치 여부: {P_B_direct == P_B_via_complement}")
print(f"분수 표현: 5/6 ≈ {5/6:.6f}")
```

```python
# 문제 2 시각화: 사건 vs 여사건
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 왼쪽: 사건 B 영역 표시
colors_event = ['#66c2a5' if f <= 5 else '#e78ac3' for f in faces]
bars1 = axes[0].bar(faces, probabilities, color=colors_event, edgecolor='white', linewidth=2, width=0.6)
axes[0].set_xlabel('주사위 눈금', fontsize=12)
axes[0].set_ylabel('확률', fontsize=12)
axes[0].set_title('방법 1: 직접 계산\nP(X≤5) = 5/6', fontsize=14, fontweight='bold')
axes[0].set_xticks(faces)
axes[0].set_ylim(0, 0.25)

# 영역 표시
axes[0].axvspan(0.5, 5.5, alpha=0.1, color='green', label=f'B = {{1,2,3,4,5}}: P={5/6:.4f}')
axes[0].legend(fontsize=10, loc='upper right')

# 오른쪽: 여사건 활용
colors_comp = ['#d9d9d9' if f <= 5 else '#fc8d62' for f in faces]
bars2 = axes[1].bar(faces, probabilities, color=colors_comp, edgecolor='white', linewidth=2, width=0.6)
axes[1].set_xlabel('주사위 눈금', fontsize=12)
axes[1].set_ylabel('확률', fontsize=12)
axes[1].set_title('방법 2: 여사건 활용\nP(X≤5) = 1 - P(X>5) = 1 - 1/6', fontsize=14, fontweight='bold')
axes[1].set_xticks(faces)
axes[1].set_ylim(0, 0.25)

axes[1].axvspan(5.5, 6.5, alpha=0.2, color='red', label=f'B$^c$ = {{6}}: P={1/6:.4f}')
axes[1].annotate('이것만 구하면\n나머지는 자동!', xy=(6, 1/6), xytext=(4.5, 0.22),
                 fontsize=11, fontweight='bold', color='#fc8d62',
                 arrowprops=dict(arrowstyle='->', color='#fc8d62', lw=2))
axes[1].legend(fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

### 해석

- **방법 1(직접 계산)**: 5개의 유리한 결과를 직접 세어 $\frac{5}{6}$을 얻었다.
- **방법 2(여사건 활용)**: "5 이하가 아닌" 경우(6만 해당)의 확률을 구해 1에서 빼면 동일한 결과를 얻는다.
- 여사건은 특히 **"적어도 하나"** 유형의 문제에서 매우 유용하다.
  - 예: "주사위를 3번 던져서 적어도 한 번은 6이 나올 확률"  
    → 직접 구하면 복잡하지만, 여사건(한 번도 6이 안 나올 확률)을 빼면 간단하다.

## 문제 3: 2개 주사위의 합이 8일 확률

### 문제 정의

공정한 주사위 2개를 동시에 던질 때, 두 눈금의 합이 8이 될 확률을 구하라.

### 수학적 풀이

**표본공간:**

두 주사위의 순서쌍으로 구성된다:

$$S = \{(i, j) \mid i, j \in \{1, 2, 3, 4, 5, 6\}\}, \quad |S| = 6 \times 6 = 36$$

**사건 정의 — 합이 8인 경우의 수 열거:**

$$C = \{(i, j) \mid i + j = 8\} = \{(2,6),\; (3,5),\; (4,4),\; (5,3),\; (6,2)\}$$

$$|C| = 5$$

**확률 계산:**

$$P(X + Y = 8) = \frac{|C|}{|S|} = \frac{5}{36} \approx 0.1389$$

> **주의**: $(3, 5)$와 $(5, 3)$은 서로 다른 근원사건이다. 첫 번째 주사위가 3이고 두 번째가 5인 경우와 그 반대는 구별된다.

```python
# 문제 3: P(X+Y=8) — 경우의 수 열거
# 표본공간 생성
two_dice_space = list(product(range(1, 7), range(1, 7)))
print(f"표본공간 크기: |S| = {len(two_dice_space)}")

# 합이 8인 경우 열거
event_sum_8 = [(d1, d2) for d1, d2 in two_dice_space if d1 + d2 == 8]
print(f"\n합이 8인 경우 (사건 C):")
for pair in event_sum_8:
    print(f"  주사위1={pair[0]}, 주사위2={pair[1]} → 합={pair[0]+pair[1]}")

P_sum_8 = len(event_sum_8) / len(two_dice_space)
print(f"\n|C| = {len(event_sum_8)}")
print(f"P(X+Y=8) = {len(event_sum_8)}/{len(two_dice_space)} = {P_sum_8:.4f}")
print(f"분수 표현: 5/36 ≈ {5/36:.6f}")
```

```python
# 6x6 합계표 생성 및 Heatmap 시각화
dice_values = range(1, 7)
sum_matrix = np.zeros((6, 6), dtype=int)

for i, d1 in enumerate(dice_values):
    for j, d2 in enumerate(dice_values):
        sum_matrix[i][j] = d1 + d2

# 합이 8인 위치를 마스크
mask_8 = (sum_matrix == 8)

fig, ax = plt.subplots(figsize=(10, 8))

# 기본 heatmap
sns.heatmap(sum_matrix, annot=True, fmt='d', cmap='YlOrRd',
            xticklabels=dice_values, yticklabels=dice_values,
            linewidths=2, linecolor='white', cbar_kws={'label': '두 주사위의 합'},
            ax=ax, annot_kws={'size': 14})

# 합이 8인 셀 강조 (빨간 테두리)
for i in range(6):
    for j in range(6):
        if mask_8[i][j]:
            ax.add_patch(plt.Rectangle((j, i), 1, 1, fill=False,
                                        edgecolor='blue', linewidth=4))

ax.set_xlabel('주사위 2', fontsize=14)
ax.set_ylabel('주사위 1', fontsize=14)
ax.set_title('두 주사위의 합 (파란 테두리 = 합이 8인 경우)\n'
             f'P(합=8) = 5/36 ≈ {5/36:.4f}', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 모든 합(2~12)에 대한 확률 분포
sum_counts = {}
for d1, d2 in two_dice_space:
    s = d1 + d2
    sum_counts[s] = sum_counts.get(s, 0) + 1

sums = sorted(sum_counts.keys())
counts = [sum_counts[s] for s in sums]
probs = [c / 36 for c in counts]

# 표 출력
df_sums = pd.DataFrame({
    '합 (X+Y)': sums,
    '경우의 수': counts,
    '확률': [f'{p:.4f}' for p in probs],
    '분수': [f'{c}/36' for c in counts]
})
print("두 주사위 합의 확률분포표:")
print("=" * 50)
print(df_sums.to_string(index=False))
print(f"\n확률의 합 = {sum(probs):.4f} (검증: 반드시 1이어야 함)")
```

```python
# 확률분포 막대그래프
fig, ax = plt.subplots(figsize=(12, 6))

colors_bar = ['#fc8d62' if s == 8 else '#66c2a5' for s in sums]
bars = ax.bar(sums, probs, color=colors_bar, edgecolor='white', linewidth=2, width=0.7)

# 경우의 수와 확률 표시
for bar, s, c, p in zip(bars, sums, counts, probs):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.003,
            f'{c}/36\n({p:.3f})', ha='center', va='bottom', fontsize=9, fontweight='bold')

ax.set_xlabel('두 주사위의 합', fontsize=14)
ax.set_ylabel('확률', fontsize=14)
ax.set_title('두 주사위 합의 확률분포 (주황 = 합이 8)', fontsize=16, fontweight='bold')
ax.set_xticks(sums)
ax.set_ylim(0, max(probs) + 0.05)

plt.tight_layout()
plt.show()
```

### 해석

- 두 주사위의 합은 2부터 12까지 가능하며, **합이 7일 때 확률이 가장 높다** ($\frac{6}{36} = \frac{1}{6}$).
- 분포가 **대칭**임을 확인할 수 있다: $P(합=k) = P(합=14-k)$
  - 예: $P(합=5) = P(합=9) = \frac{4}{36}$
- 합이 8인 경우의 5가지: $(2,6), (3,5), (4,4), (5,3), (6,2)$
- 이 분포는 **삼각분포(triangular distribution)**의 이산 버전과 유사하다.

## 시뮬레이션 검증

이론적으로 계산한 확률이 실제로 맞는지, **몬테카를로 시뮬레이션(Monte Carlo Simulation)**으로 검증한다.

> **큰 수의 법칙(Law of Large Numbers)**: 시행 횟수를 늘릴수록 상대빈도(경험적 확률)는 이론적 확률에 수렴한다.

```python
# 시뮬레이션: 주사위 2개를 10,000번 던지기
np.random.seed(42)
n_simulations = 10000

die1 = np.random.randint(1, 7, size=n_simulations)
die2 = np.random.randint(1, 7, size=n_simulations)
sums_sim = die1 + die2

# 문제 1: P(X=5) — 주사위 1개
sim_P_eq5 = np.mean(die1 == 5)
# 문제 2: P(X<=5) — 주사위 1개
sim_P_le5 = np.mean(die1 <= 5)
# 문제 3: P(X+Y=8) — 주사위 2개
sim_P_sum8 = np.mean(sums_sim == 8)

print("=" * 60)
print(f"몬테카를로 시뮬레이션 결과 (시행 횟수: {n_simulations:,}회)")
print("=" * 60)
print(f"{'문제':<20} {'이론값':<15} {'시뮬레이션':<15} {'오차':<10}")
print("-" * 60)
print(f"{'P(X=5)':<20} {1/6:<15.4f} {sim_P_eq5:<15.4f} {abs(1/6 - sim_P_eq5):<10.4f}")
print(f"{'P(X≤5)':<20} {5/6:<15.4f} {sim_P_le5:<15.4f} {abs(5/6 - sim_P_le5):<10.4f}")
print(f"{'P(X+Y=8)':<20} {5/36:<15.4f} {sim_P_sum8:<15.4f} {abs(5/36 - sim_P_sum8):<10.4f}")
```

```python
# 시뮬레이션 vs 이론값 비교 시각화
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 두 주사위 합의 상대빈도 vs 이론 확률
sim_counts = pd.Series(sums_sim).value_counts().sort_index()
sim_probs = sim_counts / n_simulations

x_vals = range(2, 13)
theoretical = [sum_counts.get(x, 0)/36 for x in x_vals]
simulated = [sim_probs.get(x, 0) for x in x_vals]

width = 0.35
x_pos = np.arange(len(x_vals))

axes[0].bar(x_pos - width/2, theoretical, width, label='이론값', color='#66c2a5', edgecolor='white')
axes[0].bar(x_pos + width/2, simulated, width, label=f'시뮬레이션 (n={n_simulations:,})', color='#fc8d62', edgecolor='white')
axes[0].set_xlabel('두 주사위의 합', fontsize=12)
axes[0].set_ylabel('확률', fontsize=12)
axes[0].set_title('이론 확률 vs 시뮬레이션 상대빈도', fontsize=14, fontweight='bold')
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(x_vals)
axes[0].legend(fontsize=11)

# 오른쪽: 큰 수의 법칙 — 시행 횟수에 따른 P(합=8) 수렴
cumulative_prob = np.cumsum(sums_sim == 8) / np.arange(1, n_simulations + 1)
axes[1].plot(range(1, n_simulations + 1), cumulative_prob, color='#8da0cb', alpha=0.8, linewidth=1)
axes[1].axhline(y=5/36, color='red', linestyle='--', linewidth=2, label=f'이론값 = 5/36 ≈ {5/36:.4f}')
axes[1].set_xlabel('시행 횟수', fontsize=12)
axes[1].set_ylabel('누적 상대빈도 P(합=8)', fontsize=12)
axes[1].set_title('큰 수의 법칙: P(합=8)의 수렴', fontsize=14, fontweight='bold')
axes[1].set_xscale('log')
axes[1].legend(fontsize=11)
axes[1].set_ylim(0, 0.3)

plt.tight_layout()
plt.show()
```

### 해석

- **왼쪽 그래프**: 10,000번 시뮬레이션의 상대빈도가 이론 확률과 매우 근접하다.
- **오른쪽 그래프**: 시행 횟수가 늘어남에 따라 경험적 확률이 이론값($\frac{5}{36}$)으로 **수렴**하는 것을 확인할 수 있다.
- 이것이 **큰 수의 법칙**의 핵심: 충분히 많은 시행을 반복하면 관측 빈도가 진확률에 가까워진다.
- 시뮬레이션은 이론적 계산의 **검증 도구**로 활용할 수 있다.

## 종합 정리

| 문제 | 사건 | 이론적 확률 | 핵심 개념 |
|------|------|-------------|----------|
| 문제 1 | $P(X=5)$ | $\frac{1}{6} \approx 0.1667$ | 고전적 확률 정의, 근원사건 |
| 문제 2 | $P(X \leq 5)$ | $\frac{5}{6} \approx 0.8333$ | 여사건, 가법성 |
| 문제 3 | $P(X+Y=8)$ | $\frac{5}{36} \approx 0.1389$ | 순서쌍, 경우의 수 열거 |

### 핵심 개념 요약

1. **표본공간**: 가능한 모든 결과의 집합. 확률 계산의 출발점.
2. **고전적 확률**: 동일 가능성 가정 하에 $P(A) = |A| / |S|$
3. **여사건**: $P(A) = 1 - P(A^c)$. 복잡한 문제를 단순화하는 강력한 도구.
4. **경우의 수 열거**: 체계적 나열로 빠짐없이 세는 것이 중요.
5. **큰 수의 법칙**: 이론과 실험의 연결고리. 시뮬레이션의 수학적 근거.
