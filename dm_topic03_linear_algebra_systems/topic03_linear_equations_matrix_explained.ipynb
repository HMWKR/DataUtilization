## 1. 연립방정식을 행렬로 표현하기

`3a + 2b = 8`, `2a - b = 3`은 `Ax=b` 형태로 쓸 수 있습니다.
행렬 형태로 바꾸면 NumPy가 해를 안정적으로 계산해 줍니다.


```python
import numpy as np

A = np.array([[3, 2], [2, -1]], dtype=float)
b = np.array([8, 3], dtype=float)

solution = np.linalg.solve(A, b)
residual = A @ solution - b

{"a": round(float(solution[0]), 3), "b": round(float(solution[1]), 3), "residual": residual.round(10).tolist()}
```

## 2. 선형독립과 행렬식

열벡터가 서로 선형독립이면 행렬식이 0이 아니고 역행렬이 존재합니다.
행렬식이 0에 가까우면 방정식의 해가 불안정해질 수 있습니다.


```python
determinant = np.linalg.det(A)
inverse = np.linalg.inv(A)
identity_check = A @ inverse

{
    "determinant": round(float(determinant), 3),
    "inverse": inverse.round(3).tolist(),
    "A_times_inverse": identity_check.round(3).tolist(),
}
```

## 3. 최소제곱으로 과잉결정 문제 풀기

데이터 분석에서는 방정식보다 관측치가 더 많은 경우가 흔합니다.
이때 모든 점을 완벽히 만족하는 해 대신 오차 제곱합이 가장 작은 해를 구합니다.


```python
import pandas as pd

hours = np.array([1, 2, 3, 4, 5], dtype=float)
score = np.array([52, 57, 65, 70, 76], dtype=float)
design = np.column_stack([np.ones_like(hours), hours])

intercept, slope = np.linalg.lstsq(design, score, rcond=None)[0]
predicted = intercept + slope * hours

pd.DataFrame(
    {
        "hours": hours,
        "score": score,
        "predicted": predicted.round(2),
        "error": (score - predicted).round(2),
    }
)
```

## 정리

- 연립방정식은 행렬식 `Ax=b`로 표현할 수 있습니다.
- 행렬식과 역행렬은 해의 존재와 안정성을 이해하는 단서입니다.
- 최소제곱은 실제 데이터에서 가장 자주 쓰이는 선형 계산 방식입니다.
