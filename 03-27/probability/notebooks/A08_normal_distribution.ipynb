**학습 목표**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import binom, bernoulli, norm, uniform, expon, pareto
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 100
sns.set_style('whitegrid')
sns.set_palette('Set2')

np.random.seed(42)
print('라이브러리 로드 완료')
```

## 1. 정규분포 $N(\mu, \sigma^2)$

### 확률밀도함수 (PDF)

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

### 핵심 매개변수

| 매개변수 | 기호 | 역할 | 영향 |
|---------|------|------|------|
| **평균** | $\mu$ | 분포의 **중심 위치** | 좌우 이동 |
| **분산** | $\sigma^2$ | 분포의 **퍼진 정도** | 폭 변화 |
| **표준편차** | $\sigma$ | $\sqrt{\sigma^2}$ | 직관적 해석 |

### 정규분포의 특성
- **대칭성**: $\mu$를 중심으로 좌우 완벽 대칭
- **단봉성**: 봉우리가 하나 ($x = \mu$에서 최대)
- **전체 면적**: $\int_{-\infty}^{\infty} f(x)dx = 1$
- **기댓값 = 중앙값 = 최빈값** = $\mu$

```python
# === 정규분포 PDF 기본 시각화 ===
x = np.linspace(-6, 6, 1000)

fig, ax = plt.subplots(figsize=(12, 7))

# 표준정규분포 N(0,1)
y = norm.pdf(x, 0, 1)
ax.plot(x, y, 'b-', linewidth=3, label='N(0, 1)')
ax.fill_between(x, y, alpha=0.1, color='blue')

# 주요 지점 표시
ax.axvline(0, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='μ = 0')

# PDF 수식 표시
ax.text(2.5, 0.35, r'$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$',
        fontsize=16, bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

# μ와 σ 위치 표시
ax.annotate('', xy=(1, 0.24), xytext=(0, 0.24),
            arrowprops=dict(arrowstyle='<->', color='green', linewidth=2))
ax.text(0.5, 0.255, 'σ = 1', ha='center', fontsize=12, color='green', fontweight='bold')

ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('f(x)', fontsize=14)
ax.set_title('정규분포의 확률밀도함수 (PDF)', fontsize=16, fontweight='bold')
ax.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

## 2. μ, σ 변경에 따른 곡선 형태 변화

```python
# === 4가지 (μ, σ) 조합 비교 ===
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
x = np.linspace(-10, 15, 1000)

# 조합 1: μ 변경 (σ 고정)
ax = axes[0, 0]
for mu_val, color in zip([-2, 0, 2, 5], sns.color_palette('husl', 4)):
    y = norm.pdf(x, mu_val, 1)
    ax.plot(x, y, '-', linewidth=2.5, color=color, label=f'N({mu_val}, 1²)')
    ax.axvline(mu_val, color=color, linestyle=':', alpha=0.3)
ax.set_title('① μ 변경, σ=1 고정\n(좌우 이동)', fontsize=13, fontweight='bold')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.legend(fontsize=10)

# 조합 2: σ 변경 (μ 고정)
ax = axes[0, 1]
for sigma_val, color in zip([0.5, 1, 2, 3], sns.color_palette('husl', 4)):
    y = norm.pdf(x, 0, sigma_val)
    ax.plot(x, y, '-', linewidth=2.5, color=color, label=f'N(0, {sigma_val}²)')
ax.set_title('② σ 변경, μ=0 고정\n(폭 변화)', fontsize=13, fontweight='bold')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.legend(fontsize=10)

# 조합 3: μ와 σ 모두 변경
ax = axes[1, 0]
params = [(0, 1, 'N(0, 1²)'), (2, 0.5, 'N(2, 0.5²)'), (-1, 2, 'N(-1, 2²)'), (3, 1.5, 'N(3, 1.5²)')]
for (mu_val, sigma_val, label), color in zip(params, sns.color_palette('husl', 4)):
    y = norm.pdf(x, mu_val, sigma_val)
    ax.plot(x, y, '-', linewidth=2.5, color=color, label=label)
ax.set_title('③ μ와 σ 모두 변경\n(위치 + 폭)', fontsize=13, fontweight='bold')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.legend(fontsize=10)

# 조합 4: 모두 표준화하면 N(0,1)
ax = axes[1, 1]
z = np.linspace(-4, 4, 1000)
original_params = [(100, 15), (50, 10), (0, 1), (170, 10)]
labels = ['시험점수 N(100,15²)', '몸무게 N(50,10²)', '표준 N(0,1²)', '키 N(170,10²)']
for (mu_val, sigma_val), label, color in zip(original_params, labels, sns.color_palette('husl', 4)):
    # 표준화하면 모두 N(0,1)
    y = norm.pdf(z, 0, 1)
ax.plot(z, norm.pdf(z, 0, 1), 'k-', linewidth=3, label='표준화 → N(0, 1)')
ax.fill_between(z, norm.pdf(z, 0, 1), alpha=0.2, color='gold')
ax.set_title('④ 모든 정규분포 → 표준화 → N(0,1)\n(하나로 통일!)', fontsize=13, fontweight='bold')
ax.set_xlabel('Z', fontsize=12)
ax.set_ylabel('f(Z)', fontsize=12)
ax.legend(fontsize=10)

plt.suptitle('정규분포: μ와 σ에 따른 형태 변화', fontsize=17, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

### 해석

- **① μ 변경**: $\mu$가 달라지면 곡선이 **좌우로 이동**한다. 형태(폭, 높이)는 변하지 않는다.
- **② σ 변경**: $\sigma$가 커지면 곡선이 **넓고 낮아**진다. $\sigma$가 작으면 **좁고 높다**. 면적(=1)은 항상 동일.
- **③ 동시 변경**: 실제 데이터는 각각 다른 $\mu$와 $\sigma$를 가진다.
- **④ 표준화의 힘**: 어떤 정규분포든 $Z = \frac{X-\mu}{\sigma}$ 변환으로 **N(0,1) 하나로 통일**할 수 있다.

> 이것이 바로 표준정규분포표 **하나만으로** 모든 정규분포의 확률을 계산할 수 있는 이유이다.

## 3. 표준정규분포 $N(0, 1)$ 변환

### 변환 공식

$$Z = \frac{X - \mu}{\sigma}$$

- $X \sim N(\mu, \sigma^2)$ → $Z \sim N(0, 1)$
- 역변환: $X = \mu + \sigma Z$

### 변환의 의미

| 원래 분포 | 변환 후 |
|-----------|--------|
| 평균 $\mu$ | 평균 **0** |
| 표준편차 $\sigma$ | 표준편차 **1** |
| 단위: cm, kg, 점 등 | **단위 없음** (σ 단위) |

```python
# === 정규분포 → 표준정규분포 변환 시각화 ===
mu_orig, sigma_orig = 170, 10  # 키 데이터

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# (1) 원래 정규분포 N(170, 10²)
x_orig = np.linspace(130, 210, 500)
y_orig = norm.pdf(x_orig, mu_orig, sigma_orig)
axes[0].plot(x_orig, y_orig, 'b-', linewidth=2.5)
axes[0].fill_between(x_orig, y_orig, alpha=0.15, color='blue')
axes[0].axvline(mu_orig, color='red', linestyle='--', linewidth=2, label=f'μ = {mu_orig}')

# σ 구간 표시
for i, (z_val, label) in enumerate([(1, '±1σ'), (2, '±2σ')]):
    axes[0].axvline(mu_orig + z_val*sigma_orig, color='green', linestyle=':', alpha=0.5)
    axes[0].axvline(mu_orig - z_val*sigma_orig, color='green', linestyle=':', alpha=0.5)
    axes[0].text(mu_orig + z_val*sigma_orig, y_orig.max() * (0.9 - i*0.15),
                 f'{mu_orig + z_val*sigma_orig}', ha='center', fontsize=10, color='green')
    axes[0].text(mu_orig - z_val*sigma_orig, y_orig.max() * (0.9 - i*0.15),
                 f'{mu_orig - z_val*sigma_orig}', ha='center', fontsize=10, color='green')

axes[0].set_xlabel('X (키, cm)', fontsize=13)
axes[0].set_ylabel('f(X)', fontsize=13)
axes[0].set_title(f'원래 분포: N({mu_orig}, {sigma_orig}²)', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)

# 변환 화살표
fig.text(0.5, 0.5, r'$Z = \frac{X - 170}{10}$', fontsize=20, ha='center',
         fontweight='bold', color='red',
         bbox=dict(boxstyle='rarrow,pad=0.3', facecolor='lightyellow', edgecolor='red'))

# (2) 표준정규분포 N(0, 1)
z = np.linspace(-4, 4, 500)
y_std = norm.pdf(z, 0, 1)
axes[1].plot(z, y_std, 'r-', linewidth=2.5)
axes[1].fill_between(z, y_std, alpha=0.15, color='red')
axes[1].axvline(0, color='red', linestyle='--', linewidth=2, label='μ = 0')

# σ 구간 표시
for z_val in [-2, -1, 1, 2]:
    axes[1].axvline(z_val, color='green', linestyle=':', alpha=0.5)
    axes[1].text(z_val, y_std.max() * 0.9, f'{z_val}σ', ha='center', fontsize=10, color='green')

axes[1].set_xlabel('Z (표준편차 단위)', fontsize=13)
axes[1].set_ylabel('f(Z)', fontsize=13)
axes[1].set_title('표준정규분포: N(0, 1)', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)

plt.suptitle('정규분포 → 표준정규분포 변환', fontsize=17, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# 대응 관계
print('=== X ↔ Z 대응 관계 ===')
x_values = [140, 150, 160, 170, 180, 190, 200]
for x_val in x_values:
    z_val = (x_val - mu_orig) / sigma_orig
    print(f'  X = {x_val}cm  →  Z = ({x_val} - {mu_orig}) / {sigma_orig} = {z_val:+.1f}')
```

### 해석

- **왼쪽**: 키 데이터의 원래 정규분포. 단위가 cm이고, 중심이 170에 위치한다.
- **오른쪽**: 표준정규분포로 변환된 결과. 단위가 없고, 중심이 0에 위치한다.
- 예: 키 180cm = Z=+1 (평균보다 1 표준편차 위), 키 150cm = Z=-2 (평균보다 2 표준편차 아래)
- 변환 전후에 **확률(면적)은 보존**된다. 형태도 동일하고, 스케일과 위치만 바뀐다.

## 4. σ-구간 확률 유도

### 4.1 $P(-1 < Z \leq 1) = 0.6827 \approx 0.68$

```python
# === P(-1 < Z ≤ 1) = 0.6827 유도 ===
print('=== P(-1 < Z ≤ 1) 산술식 유도 ===')
print()

cdf_1 = norm.cdf(1)      # Φ(1)
cdf_neg1 = norm.cdf(-1)  # Φ(-1)

print(f'① Φ(1)  = P(Z ≤ 1)  = norm.cdf(1)  = {cdf_1:.6f}')
print(f'② Φ(-1) = P(Z ≤ -1) = norm.cdf(-1) = {cdf_neg1:.6f}')
print()
print(f'③ P(-1 < Z ≤ 1) = Φ(1) - Φ(-1)')
print(f'                 = {cdf_1:.6f} - {cdf_neg1:.6f}')
print(f'                 = {cdf_1 - cdf_neg1:.6f}')
print(f'                 ≈ 0.68 (68%)')

# 시각화
fig, ax = plt.subplots(figsize=(12, 7))
z = np.linspace(-4, 4, 1000)
y = norm.pdf(z, 0, 1)

ax.plot(z, y, 'k-', linewidth=2)

# ±1σ 영역 색칠
z_fill = np.linspace(-1, 1, 500)
ax.fill_between(z_fill, norm.pdf(z_fill, 0, 1), alpha=0.4,
                color=sns.color_palette('Set2')[0], label=f'P(-1 < Z ≤ 1) = {cdf_1-cdf_neg1:.4f}')

# 좌측 꼬리
z_left = np.linspace(-4, -1, 300)
ax.fill_between(z_left, norm.pdf(z_left, 0, 1), alpha=0.2,
                color='gray', label=f'P(Z ≤ -1) = {cdf_neg1:.4f}')

# 우측 꼬리
z_right = np.linspace(1, 4, 300)
ax.fill_between(z_right, norm.pdf(z_right, 0, 1), alpha=0.2,
                color='gray', label=f'P(Z > 1) = {1-cdf_1:.4f}')

# 경계선
ax.axvline(-1, color='red', linestyle='--', linewidth=2)
ax.axvline(1, color='red', linestyle='--', linewidth=2)

# 확률 텍스트
ax.text(0, 0.15, f'68.27%', ha='center', fontsize=18, fontweight='bold',
        color=sns.color_palette('Set2')[0])
ax.text(-2.5, 0.03, f'{cdf_neg1:.2%}', ha='center', fontsize=12, color='gray')
ax.text(2.5, 0.03, f'{1-cdf_1:.2%}', ha='center', fontsize=12, color='gray')

# 산술식 표시
ax.text(3.5, 0.35, f'Φ(1) = {cdf_1:.4f}\nΦ(-1) = {cdf_neg1:.4f}\n\n'
        f'P = {cdf_1:.4f} - {cdf_neg1:.4f}\n  = {cdf_1-cdf_neg1:.4f}',
        fontsize=11, bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8),
        verticalalignment='top')

ax.set_xlabel('Z', fontsize=14)
ax.set_ylabel('f(Z)', fontsize=14)
ax.set_title('P(-1 < Z ≤ 1) = 0.6827 ≈ 68%', fontsize=16, fontweight='bold')
ax.legend(fontsize=11, loc='upper left')
plt.tight_layout()
plt.show()
```

### 해석

**산술식 유도**:

$$P(-1 < Z \leq 1) = \Phi(1) - \Phi(-1) = 0.8413 - 0.1587 = 0.6827$$

- $\Phi(z)$는 표준정규분포의 **누적분포함수(CDF)**: $\Phi(z) = P(Z \leq z)$
- 정규분포의 대칭성에 의해: $\Phi(-z) = 1 - \Phi(z)$
- 따라서: $P(-1 < Z \leq 1) = \Phi(1) - (1 - \Phi(1)) = 2\Phi(1) - 1 = 2(0.8413) - 1 = 0.6827$

> **의미**: 정규분포를 따르는 데이터의 약 **68%**가 평균 ±1 표준편차 범위 안에 들어간다.

### 4.2 $P(-2 < Z \leq 2)$ 와 $P(-3 < Z \leq 3)$

```python
# === ±2σ, ±3σ 구간 유도 ===
print('=== σ-구간 확률 산술식 ===')
print()

for z_val in [1, 2, 3]:
    cdf_pos = norm.cdf(z_val)
    cdf_neg = norm.cdf(-z_val)
    prob = cdf_pos - cdf_neg
    
    print(f'--- ±{z_val}σ ---')
    print(f'  Φ({z_val})  = {cdf_pos:.6f}')
    print(f'  Φ(-{z_val}) = {cdf_neg:.6f}')
    print(f'  P(-{z_val} < Z ≤ {z_val}) = {cdf_pos:.6f} - {cdf_neg:.6f} = {prob:.6f} ≈ {prob:.3f}')
    print()

# 3개 구간을 하나의 그래프로
fig, ax = plt.subplots(figsize=(14, 7))
z = np.linspace(-4, 4, 1000)
y = norm.pdf(z, 0, 1)

ax.plot(z, y, 'k-', linewidth=2.5)

# 3σ (가장 넓은 것부터)
colors = [sns.color_palette('Set2')[i] for i in [2, 1, 0]]
labels_info = [
    (3, '±3σ = 99.73%', colors[0]),
    (2, '±2σ = 95.45%', colors[1]),
    (1, '±1σ = 68.27%', colors[2])
]

for z_val, label, color in labels_info:
    z_fill = np.linspace(-z_val, z_val, 500)
    ax.fill_between(z_fill, norm.pdf(z_fill, 0, 1), alpha=0.25, color=color, label=label)
    ax.axvline(-z_val, color='gray', linestyle=':', alpha=0.4)
    ax.axvline(z_val, color='gray', linestyle=':', alpha=0.4)

# 각 영역에 퍼센트 표기
ax.text(0, 0.05, '68.27%', ha='center', fontsize=14, fontweight='bold', color=colors[2])
ax.text(0, 0.01, '95.45%', ha='center', fontsize=12, fontweight='bold', color=colors[1])

# x축 레이블
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
ax.set_xticklabels(['-3σ', '-2σ', '-1σ', '0', '+1σ', '+2σ', '+3σ'], fontsize=12)

ax.set_xlabel('Z (표준편차 단위)', fontsize=14)
ax.set_ylabel('f(Z)', fontsize=14)
ax.set_title('68 - 95 - 99.7 법칙 (경험적 법칙)', fontsize=16, fontweight='bold')
ax.legend(fontsize=12, loc='upper right')
plt.tight_layout()
plt.show()
```

### 해석

**68-95-99.7 법칙** (경험적 법칙, Empirical Rule):

| 구간 | 확률 | 의미 |
|------|------|------|
| $\pm 1\sigma$ | 68.27% | 약 $\frac{2}{3}$의 데이터 |
| $\pm 2\sigma$ | 95.45% | 거의 모든 데이터 |
| $\pm 3\sigma$ | 99.73% | 극단값 제외 사실상 전부 |

## 5. $P(-1.96 < Z \leq 1.96) = 0.95$ 정확한 유도

### 왜 1.96인가?

"정확히 95%"를 포함하는 구간을 찾고 싶다면:
- $\pm 2\sigma$는 95.45%로 약간 많다
- 정확히 95%가 되는 Z값을 찾으면 → **Z = ±1.96**

```python
# === P(-1.96 < Z ≤ 1.96) = 0.95 정확한 유도 ===
print('=== P(-1.96 < Z ≤ 1.96) = 0.95 산술식 유도 ===')
print()

cdf_196 = norm.cdf(1.96)
cdf_neg196 = norm.cdf(-1.96)

print('[ 단계 1: CDF 값 계산 ]')
print(f'  Φ(1.96)  = norm.cdf(1.96)  = {cdf_196:.6f}')
print(f'  Φ(-1.96) = norm.cdf(-1.96) = {cdf_neg196:.6f}')
print()

print('[ 단계 2: 구간 확률 계산 ]')
print(f'  P(-1.96 < Z ≤ 1.96) = Φ(1.96) - Φ(-1.96)')
print(f'                       = {cdf_196:.6f} - {cdf_neg196:.6f}')
print(f'                       = {cdf_196 - cdf_neg196:.6f}')
print(f'                       ≈ 0.95 (95%)')
print()

print('[ 단계 3: 대칭성 이용한 간편 유도 ]')
print(f'  Φ(1.96) = 0.975 (정확히!)')
print(f'  대칭성: Φ(-1.96) = 1 - Φ(1.96) = 1 - 0.975 = 0.025')
print(f'  P = 0.975 - 0.025 = 0.95')
print()

print('[ 핵심 ] 95% 신뢰구간의 기원!')
print(f'  → 양쪽 꼬리에 각각 2.5%씩 남기면 정확히 95%가 중앙에 포함된다')
print(f'  → 이것이 "95% 신뢰구간"에서 Z = 1.96을 사용하는 이유이다!')

# 역으로 확인: 95%가 되는 Z값은?
z_95 = norm.ppf(0.975)  # 상위 2.5% 지점
print(f'\n[ 검증 ] norm.ppf(0.975) = {z_95:.6f} ≈ 1.96')
```

```python
# === P(-1.96 < Z ≤ 1.96) = 0.95 시각화 ===
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

z = np.linspace(-4, 4, 1000)
y = norm.pdf(z, 0, 1)

# (1) Shaded area plot
ax = axes[0]
ax.plot(z, y, 'k-', linewidth=2.5)

# 95% 중앙 영역
z_center = np.linspace(-1.96, 1.96, 500)
ax.fill_between(z_center, norm.pdf(z_center, 0, 1), alpha=0.4,
                color=sns.color_palette('Set2')[0],
                label=f'P(-1.96 < Z ≤ 1.96) = 0.9500')

# 좌측 꼬리 (2.5%)
z_left = np.linspace(-4, -1.96, 300)
ax.fill_between(z_left, norm.pdf(z_left, 0, 1), alpha=0.4,
                color='red', label='좌측 꼬리 = 0.0250')

# 우측 꼬리 (2.5%)
z_right = np.linspace(1.96, 4, 300)
ax.fill_between(z_right, norm.pdf(z_right, 0, 1), alpha=0.4,
                color='red', label='우측 꼬리 = 0.0250')

# 경계선
ax.axvline(-1.96, color='darkred', linestyle='--', linewidth=2)
ax.axvline(1.96, color='darkred', linestyle='--', linewidth=2)

# 확률 텍스트
ax.text(0, 0.18, '95.00%', ha='center', fontsize=20, fontweight='bold',
        color=sns.color_palette('Set2')[0])
ax.text(-3, 0.02, '2.50%', ha='center', fontsize=13, fontweight='bold', color='red')
ax.text(3, 0.02, '2.50%', ha='center', fontsize=13, fontweight='bold', color='red')

# Z값 표시
ax.text(-1.96, -0.03, 'Z = -1.96', ha='center', fontsize=11, fontweight='bold', color='darkred')
ax.text(1.96, -0.03, 'Z = 1.96', ha='center', fontsize=11, fontweight='bold', color='darkred')

ax.set_xlabel('Z', fontsize=14)
ax.set_ylabel('f(Z)', fontsize=14)
ax.set_title('P(-1.96 < Z ≤ 1.96) = 0.95\n— 95% 신뢰구간의 기원', fontsize=15, fontweight='bold')
ax.legend(fontsize=11, loc='upper right')

# (2) CDF로 확인
ax2 = axes[1]
z_cdf = np.linspace(-4, 4, 1000)
y_cdf = norm.cdf(z_cdf, 0, 1)

ax2.plot(z_cdf, y_cdf, 'b-', linewidth=2.5, label='Φ(z) = CDF')

# Φ(1.96) = 0.975
ax2.plot(1.96, 0.975, 'ro', markersize=10, zorder=5)
ax2.annotate(f'Φ(1.96) = 0.975', xy=(1.96, 0.975), xytext=(2.8, 0.85),
             fontsize=12, fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='red', linewidth=2))

# Φ(-1.96) = 0.025
ax2.plot(-1.96, 0.025, 'ro', markersize=10, zorder=5)
ax2.annotate(f'Φ(-1.96) = 0.025', xy=(-1.96, 0.025), xytext=(-3.5, 0.15),
             fontsize=12, fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='red', linewidth=2))

# 차이 = 0.95
ax2.annotate('', xy=(3.8, 0.975), xytext=(3.8, 0.025),
             arrowprops=dict(arrowstyle='<->', color='green', linewidth=3))
ax2.text(3.95, 0.5, '0.975 - 0.025\n= 0.950', fontsize=13, fontweight='bold',
         color='green', va='center')

# 보조선
ax2.axhline(0.975, color='red', linestyle=':', alpha=0.5)
ax2.axhline(0.025, color='red', linestyle=':', alpha=0.5)
ax2.axvline(-1.96, color='red', linestyle=':', alpha=0.5)
ax2.axvline(1.96, color='red', linestyle=':', alpha=0.5)

ax2.set_xlabel('Z', fontsize=14)
ax2.set_ylabel('Φ(Z)', fontsize=14)
ax2.set_title('누적분포함수(CDF)로 확인\nΦ(1.96) - Φ(-1.96) = 0.95', fontsize=15, fontweight='bold')
ax2.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

### 해석

**$P(-1.96 < Z \leq 1.96) = 0.95$ 산술식**:

$$P(-1.96 < Z \leq 1.96) = \Phi(1.96) - \Phi(-1.96)$$

$$= 0.9750 - 0.0250 = 0.9500$$

**핵심 포인트**:
- $\Phi(1.96) = 0.975$ → Z=1.96 이하에 전체의 97.5%가 포함
- $\Phi(-1.96) = 0.025$ → Z=-1.96 이하에 전체의 2.5%만 포함
- 따라서 중앙에 $97.5\% - 2.5\% = 95\%$가 포함
- 양쪽 꼬리에 각각 $2.5\%$씩 남김 → 이것이 **95% 신뢰구간**의 수학적 기원!

> **통계적 추론**: 모평균의 95% 신뢰구간은 $\bar{X} \pm 1.96 \cdot \frac{\sigma}{\sqrt{n}}$이며, 여기서 1.96이 바로 이 값이다.

## 6. 표준정규분포표 만들기

```python
# === 표준정규분포표: Z=0.00 ~ 3.99 ===
# 행: Z의 정수+소수 첫째자리 (0.0, 0.1, ..., 3.9)
# 열: Z의 소수 둘째자리 (0.00, 0.01, ..., 0.09)

rows = np.arange(0.0, 4.0, 0.1)  # 0.0, 0.1, ..., 3.9
cols = np.arange(0.00, 0.10, 0.01)  # 0.00, 0.01, ..., 0.09

# 분포표 생성
z_table_data = {}
for col_val in cols:
    col_name = f'.0{int(col_val*100)}' if col_val < 0.095 else f'.{int(col_val*100)}'
    col_name = f'{col_val:.2f}'[1:]  # .00, .01, ..., .09
    values = []
    for row_val in rows:
        z = round(row_val + col_val, 2)
        values.append(round(norm.cdf(z), 4))
    z_table_data[col_name] = values

z_table = pd.DataFrame(z_table_data, index=[f'{r:.1f}' for r in rows])
z_table.index.name = 'Z'

print('=== 표준정규분포표 Φ(z) = P(Z ≤ z) ===')
print('  (행: Z의 첫째 소수, 열: Z의 둘째 소수)')
print()
print(z_table.to_string())
```

```python
# === 분포표에서 주요 값 확인 ===
print('=== 표준정규분포표에서 주요 Z값 확인 ===')
print()

key_values = [
    (0.00, '평균 (50% 지점)'),
    (1.00, '±1σ 경계 (84.13%)'),
    (1.64, '90% 신뢰구간 (편측 5%)'),
    (1.96, '95% 신뢰구간 (편측 2.5%)'),
    (2.00, '±2σ 경계 (97.72%)'),
    (2.33, '98% 신뢰구간 (편측 1%)'),
    (2.58, '99% 신뢰구간 (편측 0.5%)'),
    (3.00, '±3σ 경계 (99.87%)'),
    (3.29, '99.9% 신뢰구간'),
]

for z_val, meaning in key_values:
    cdf_val = norm.cdf(z_val)
    print(f'  Φ({z_val:.2f}) = {cdf_val:.4f}  →  {meaning}')
```

```python
# === 분포표 히트맵 시각화 ===
fig, ax = plt.subplots(figsize=(14, 16))

# Z=0.0~1.9 부분만 히트맵으로 (가독성)
z_table_subset = z_table.iloc[:20]  # 0.0 ~ 1.9
z_values_float = z_table_subset.values.astype(float)

heatmap = sns.heatmap(z_values_float, annot=True, fmt='.4f', cmap='YlOrRd',
                       linewidths=0.5, linecolor='white',
                       xticklabels=z_table_subset.columns,
                       yticklabels=z_table_subset.index,
                       cbar_kws={'label': 'Φ(z)'}, ax=ax,
                       annot_kws={'fontsize': 8})

ax.set_xlabel('Z 소수 둘째자리', fontsize=13)
ax.set_ylabel('Z 값', fontsize=13)
ax.set_title('표준정규분포표 히트맵 (Z = 0.0 ~ 1.9)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 해석

- **표준정규분포표**는 $\Phi(z) = P(Z \leq z)$의 값을 미리 계산해놓은 표이다.
- Z=0.00에서 $\Phi(0) = 0.5000$ (전체의 정확히 50%)
- Z=1.96에서 $\Phi(1.96) = 0.9750$ (전체의 97.5%)
- Z가 커질수록 $\Phi(z)$는 1에 수렴한다.
- 히트맵의 색이 진해질수록(빨간색) $\Phi(z)$ 값이 1에 가깝다.

> **현대에는** Python의 `norm.cdf()`로 즉시 계산하지만, 분포표를 이해하는 것은 확률 계산의 원리를 파악하는 데 필수적이다.

```python
# === 주요 확률 구간 종합 시각화 ===
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
z = np.linspace(-4, 4, 1000)
y = norm.pdf(z, 0, 1)

# 시각화할 구간 목록
intervals = [
    (-1, 1, '±1σ', f'{norm.cdf(1)-norm.cdf(-1):.4f}', '68.27%'),
    (-2, 2, '±2σ', f'{norm.cdf(2)-norm.cdf(-2):.4f}', '95.45%'),
    (-3, 3, '±3σ', f'{norm.cdf(3)-norm.cdf(-3):.4f}', '99.73%'),
    (-1.96, 1.96, '±1.96σ', f'{norm.cdf(1.96)-norm.cdf(-1.96):.4f}', '95.00%'),
    (-2.576, 2.576, '±2.576σ', f'{norm.cdf(2.576)-norm.cdf(-2.576):.4f}', '99.00%'),
    (0, 4, 'Z > 0', f'{1-norm.cdf(0):.4f}', '50.00%')
]

colors = sns.color_palette('Set2', 6)

for idx, (z_low, z_high, label, prob_str, pct) in enumerate(intervals):
    ax = axes[idx // 3][idx % 3]
    ax.plot(z, y, 'k-', linewidth=2)
    
    # 색칠 영역
    z_fill = np.linspace(z_low, z_high, 500)
    ax.fill_between(z_fill, norm.pdf(z_fill, 0, 1), alpha=0.4, color=colors[idx])
    
    # 경계선
    if z_low > -3.5:
        ax.axvline(z_low, color='red', linestyle='--', linewidth=1.5)
    if z_high < 3.5:
        ax.axvline(z_high, color='red', linestyle='--', linewidth=1.5)
    
    # 확률 텍스트
    ax.text((z_low+z_high)/2, 0.15, f'{pct}', ha='center', fontsize=16, fontweight='bold',
            color=colors[idx])
    
    # 산술식
    ax.text(0, 0.42, f'Φ({z_high:.2f}) - Φ({z_low:.2f})\n= {prob_str}',
            ha='center', fontsize=10,
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))
    
    ax.set_xlabel('Z', fontsize=11)
    ax.set_ylabel('f(Z)', fontsize=11)
    ax.set_title(f'{label} 구간', fontsize=13, fontweight='bold')
    ax.set_xlim(-4, 4)
    ax.set_ylim(-0.02, 0.45)

plt.suptitle('표준정규분포 주요 확률 구간 종합', fontsize=17, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

> 이 6개 구간은 통계학과 머신러닝에서 가장 빈번하게 사용되는 확률 구간이다.

## 8. 정리

### 핵심 공식

| 항목 | 공식/값 |
|------|--------|
| 정규분포 PDF | $f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$ |
| 표준화 변환 | $Z = \frac{X - \mu}{\sigma}$ |
| $P(-1 < Z \leq 1)$ | $\Phi(1) - \Phi(-1) = 0.8413 - 0.1587 = \mathbf{0.6827}$ |
| $P(-1.96 < Z \leq 1.96)$ | $\Phi(1.96) - \Phi(-1.96) = 0.9750 - 0.0250 = \mathbf{0.9500}$ |
| $P(-2 < Z \leq 2)$ | $0.9772 - 0.0228 = 0.9545$ |
| $P(-3 < Z \leq 3)$ | $0.9987 - 0.0013 = 0.9973$ |

### 핵심 요약

1. **정규분포** $N(\mu, \sigma^2)$: $\mu$는 중심 위치, $\sigma$는 퍼진 정도를 결정한다.
2. **표준정규분포** $N(0, 1)$: 모든 정규분포는 $Z = (X-\mu)/\sigma$ 변환으로 하나의 분포표로 통일된다.
3. **68-95-99.7 법칙**: $\pm 1\sigma$에 68%, $\pm 2\sigma$에 95%, $\pm 3\sigma$에 99.7%.
4. **95% 신뢰구간**: $Z = \pm 1.96$에서 정확히 95%. $\Phi(1.96) = 0.975$이 핵심.
5. **표준정규분포표**: $\Phi(z) = P(Z \leq z)$를 미리 계산한 참조 테이블.
