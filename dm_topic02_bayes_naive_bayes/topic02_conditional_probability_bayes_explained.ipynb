## 1. 조건부확률 계산

조건부확률 `P(A|B)`는 B가 일어났다는 조건 아래 A가 일어날 확률입니다.
작은 고객 데이터로 구매 여부와 멤버십 등급의 관계를 확인합니다.


```python
import pandas as pd

customers = pd.DataFrame(
    {
        "grade": ["basic", "silver", "gold", "gold", "silver", "basic", "gold", "silver"],
        "used_coupon": [0, 1, 1, 1, 0, 0, 1, 1],
        "purchased": [0, 1, 1, 1, 0, 0, 1, 0],
    }
)

p_purchase_given_gold = customers.query("grade == 'gold'")["purchased"].mean()
p_coupon_given_purchase = customers.query("purchased == 1")["used_coupon"].mean()

{
    "P(purchase | gold)": round(float(p_purchase_given_gold), 3),
    "P(coupon | purchase)": round(float(p_coupon_given_purchase), 3),
}
```

## 2. 베이즈 정리로 사후확률 구하기

베이즈 정리는 원인에 대한 믿음인 사전확률을 관측 증거로 갱신합니다.
예를 들어 검사 결과가 양성일 때 실제로 질병이 있을 확률을 계산할 수 있습니다.


```python
def bayes(prior, sensitivity, false_positive_rate):
    evidence = sensitivity * prior + false_positive_rate * (1 - prior)
    posterior = sensitivity * prior / evidence
    return posterior


disease_prior = 0.02
sensitivity = 0.95
false_positive_rate = 0.08
posterior = bayes(disease_prior, sensitivity, false_positive_rate)

print(f"검사 양성일 때 실제 질병일 확률: {posterior:.3f}")
```

## 3. Gaussian Naive Bayes 분류

나이브 베이즈는 특징들이 조건부 독립이라고 단순화합니다.
이 가정은 강하지만, 작은 데이터에서도 빠르게 기준 모델을 만들 수 있습니다.


```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

X = np.array(
    [
        [1, 18],
        [1, 22],
        [0, 30],
        [0, 35],
        [1, 28],
        [0, 40],
        [1, 24],
        [0, 45],
    ]
)
y = np.array(["buy", "buy", "skip", "skip", "buy", "skip", "buy", "skip"])

model = GaussianNB()
model.fit(X, y)
prediction = model.predict([[1, 26]])[0]
probability = model.predict_proba([[1, 26]])[0].max()

{"prediction": prediction, "confidence": round(float(probability), 3)}
```

- 베이즈 정리는 증거가 들어왔을 때 사후확률을 갱신합니다.
- 나이브 베이즈는 빠른 분류 기준 모델로 유용합니다.
